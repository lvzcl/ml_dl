# 步骤如下
    1.获取inputs
    2.构建generator
    3.构建discrimitor
    4.定义参数
    5.loss&optimizer
    6.训练模型
    7.显示结果

```python
#1.get inputs
def get_inputs(real_size, noise_size):
    #real_images是作为discrimitor的输入
    #noise_images是作为generator的输入，生成对应的image
    real_images = tf.placeholder(tf.float32, [None, real_size], name='real_images')
    noise_images = tf.placeholder(tf.float32, [None, noise_size], name='noise_size')
    return real_images, noise_images
    
```

```python
#2.构建generator,一个dnn网络，一个隐藏层，当然也可以采用卷积网络
def generator(noise_images, g_units, out_dim, reuse=False, alpha=0.01):
    #隐层使用leaky_relu,alpha 为0.01，输出层使用tanh，参考的文献说tanh的效果比sigmoid的效果好
    with tf.variable_scope('generator', reuse=reuse):
        hidden = tf.layers.dense(noise_images, g_units)
        hidden = tf.maximum(alpha*hidden ,hidden)
        dropout = tf.layers.dropout(hidden, rate=0.2)
        logits = tf.layers.dense(dropout, out_dim)
        output = tf.tanh(logits)
        return logits, output
        
```

```python
# 构造判别器
def discrimitor(input_img, d_units, reuse=False, alpha=0.01):
    with tf.variable_scope('discrimitor', reuse=reuse):
        hidden = tf.layers.dense(input_img, d_units)
        hidden = tf.maximum(alpha*hidden, hidden)
        logits = tf.layers.dense(hidden, 1)
        output = tf.sigmoid(logits)
        return logits, output
```

```python
#参数设置
image_size = train_data.images[0].shape[0]
noise_size = 100
g_units = 128
d_units = 128
alpha = 0.01
learning_rate = 0.001
smooth = 0.1
```

```python
#接下来就是网络的构建
tf.reset_default_graph()
real_images, noise_images = get_inputs(image_size, noise_size)
#将noise_images放入generator
generator_logits, generator_output = generator(noise_images, g_units, image_size)
d_logits_real, d_output_real = discrimitor(real_images, d_units)
d_logits_fake, d_output_fake = discrimitor(generator_output, d_units, reuse=True)
```

```python
#loss 和 optimizer
d_loss_real = tf.reduce_mean(tf.nn.sigmoid_cross_entropy_with_logits(logits=d_logits_real, 
                                                                     labels=tf.ones_like(d_logits_real)))
d_loss_fake = tf.reduce_mean(tf.nn.sigmoid_cross_entropy_with_logits(logits=d_logits_fake, 
                                                                     labels=tf.zeros_like(d_logits_fake)))
d_loss = tf.add(d_loss_real, d_loss_fake)

g_loss = tf.reduce_mean(tf.nn.sigmoid_cross_entropy_with_logits(logits=d_logits_fake, 
                                                                labels=tf.ones_like(d_logits_fake)))
```

```python
train_vars = tf.trainable_variables()
g_vars = [var for var in train_vars if var.name.startswith('generator')]
d_vars = [var for var in train_vars if var.name.startswith('discrimitor')]

d_optimizer = tf.train.AdamOptimizer(learning_rate).minimize(d_loss, var_list = d_vars)
g_optimizer = tf.train.AdamOptimizer(learning_rate).minimize(g_loss, var_list = g_vars)
```

```python
batch_size = 64
epochs = 300
n_samples = 25
#存储测试样例
samples = []
#存储loss
losses = []
#存储model
saver = tf.train.Saver(var_list=g_vars)
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    for epoch in range(epochs):
        for batch_i in range(train_data.num_examples // batch_size):
            batch = train_data.next_batch(batch_size)
            batch_images = batch[0].reshape((batch_size, 784))
            #因为使用了tanh，取值范围在(-1,1)
            batch_image = 2*batch_images - 1
            
            batch_noise = np.random.uniform(-1, 1, size=(batch_size,noise_size))
            
            _ = sess.run(d_optimizer, feed_dict={real_images:batch_images, noise_images:batch_noise})
            _ = sess.run(g_optimizer, feed_dict={noise_images:batch_noise})
        #计算loss
        train_d_loss = sess.run(d_loss, feed_dict = {real_images:batch_images, noise_images:batch_noise})
        train_d_real_loss = sess.run(d_loss_real, feed_dict = {real_images:batch_images})
        train_d_fake_loss = sess.run(d_loss_fake, feed_dict = {noise_images:batch_noise})
        train_g_loss = sess.run(g_loss, feed_dict = {noise_images:batch_noise})
        print("Epoch {}/{}...".format(epoch+1, epochs),
              "Discriminator Loss: {:.4f}(Real: {:.4f} + Fake: {:.4f})...".format(train_d_loss, train_d_real_loss, train_d_fake_loss),
              "Generator Loss: {:.4f}".format(train_g_loss))
        losses.append((train_d_loss, train_d_real_loss, train_d_fake_loss, train_g_loss))
        #抽取样本，后期测试
        sample_noise = np.random.uniform(-1, 1, size = (n_samples, noise_size))
        gen_samples = sess.run(generator(noise_images, g_units, image_size, reuse=True), 
                               feed_dict={noise_images:sample_noise})
        samples.append(gen_samples)
        temp = saver.save(sess, './model/generator.ckpt')
        
```

```python
def view_samples(epoch, samples):
    """
    epoch代表第几次迭代的图像
    samples为我们的采样结果
    """
    fig, axes = plt.subplots(figsize=(7,7), nrows=5, ncols=5, sharey=True, sharex=True)
    for ax, img in zip(axes.flatten(), samples[epoch][1]): # 这里samples[epoch][1]代表生成的图像结果，而[0]代表对应的logits
        ax.xaxis.set_visible(False)
        ax.yaxis.set_visible(False)
        im = ax.imshow(img.reshape((28,28)), cmap='Greys_r')
    plt.show()
    return fig, axes
```
